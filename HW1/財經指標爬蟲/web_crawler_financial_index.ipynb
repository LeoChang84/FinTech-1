```python
import csv
from io import BytesIO
import pandas as pd
import requests
from bs4 import BeautifulSoup
#瀏覽器的規格
head={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'}
#下載到最新的Industrial Production and Capacity Utilization的csv檔案
url = 'https://www.federalreserve.gov/datadownload/Output.aspx?rel=G17&series=809009461b1cba2fd5b4cf5557d9d663&lastobs=&from=&to=&filetype=csv&label=include&layout=seriesrow&type=package'
#把下載的csv檔案存在download內
download=requests.get(url, headers=head)
#取出來存成dataframe的資料格式
df =  pd.read_csv(BytesIO(download.content))
#轉置表格
df = df.transpose()
#第一列設為columns
df = df.set_axis(df.iloc[0], axis='columns', inplace=False)
#刪掉前面不是時間列的資料列
df = df.drop(df.index[:6], axis=0)
#將index的名稱設為Date
df.index.names = ['Date']
df
#下表呈現很多工業生產指數，但總指數為Total index; s.a. IP欄位，因此要將該欄位取出
```
